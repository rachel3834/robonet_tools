# Image Quality Control

LCO 1m imaging data tends to suffer from structure in the sky background of images when the Moon is in close proximity.  
The structure can take a range of forms from a linear 2D gradient to a complex cross-grid pattern spanning the whole image.
Good photometry from these data using a DIA pipeline is rare, since the structure varies with time (i.e. Moon position) and does not subtract cleanly.  

The goal of this notebook is to identifiy an algorithm which can detect these frames and flag them so that the automatic pipeline removes them from the data stream.  

## Current pipeline quality control

Stage 1 performs an initial evaluation of image quality, based on the number of stars detected, sky background statistics and PSF statistics, based on DAOphot results from a small section in the center of the frame. 

A few issues with this approach have become apparent in real-world operations:
* The number of stars detected in the center of the frame can drop below the selection threshold for targets that are outside the Galactic Plane.  This is a false-positive rejection because these fields will always be low stellar density.
* Structure in the image background manifests over a wide area than the image center.  The image statistics from this area alone make it hard to identify images with high or structured sky background

## Examining the pipeline's current default algorithm

As a baseline for comparision, let's run the pipeline's code to measure the sky background and other quality control statistics with the default configuration for the test of test images. 

```python
# Set up configuration parameters required by pyDANDIA
red_params = {
    'red_dir': data_dir,
    'db_file_path': None,
    'mode': 'auto',
    'primary_flag': 'primary_ref',
    'build_phot_db': False,
    'log_dir': data_dir,
    'pipeline_config_dir': '/User/rstreet/OMEGAII/configs/',
    'base_dir': data_dir,
    'software_dir': '/User/rstreet/software/pyDANDIA/pyDANDIA/'
}
setup = pipeline_setup.pipeline_setup(red_params)
reduction_metadata = metadata.MetaData()
reduction_metadata.load_all_metadata(metadata_directory=data_dir,
                                             metadata_name='pyDANDIA_metadata.fits')
log = logs.start_stage_log(setup.red_dir, 'test')
```

```python
def time_diff(t1, t2):
    dt = t2 - t1
    delta = dt.seconds + dt.microseconds/1000
    return delta
```

```python
def init_results():
    results = {
        'sky': np.zeros(len(image_list)),
        'sky_sigma': np.zeros(len(image_list)),
        'sigma_y': np.zeros(len(image_list)),
        'sigma_x': np.zeros(len(image_list)),
        'corr_xy': np.zeros(len(image_list)),
        'nstars': np.zeros(len(image_list)),
        'sat_frac': np.zeros(len(image_list)),
        'symmetry': np.zeros(len(image_list))
    }
    return results
```

```python
# Run the DAOphot starfind, as performed by pyDANDIA.stage1, on the test images
results = init_results()
total_time = 0
for i,image in enumerate(image_list):
    t1 = datetime.now(UTC)
    (status, report, params) = starfind.starfind(setup, image, reduction_metadata,
                                                     plot_it=False, log=log, thumbsize = 500)
    t2 = datetime.now(UTC)
    dt = time_diff(t1, t2)
    print('Time taken = ' + repr(dt)+'s')
    total_time += dt
    for metric, value in params.items():
        results[metric][i] = value

print('Total time taken = ' + repr(total_time) + 's; ' + str(total_time/len(results))+'s per frame')
print(results)
```

Now we can compare the quality statistics if we use the full frame images

```python
full_frame_results = init_results()
total_time = 0
for i,image in enumerate(image_list):
    t1 = datetime.now(UTC)
    (status, report, params) = starfind.starfind(setup, image, reduction_metadata,
                                                     plot_it=False, log=log, thumbsize = 4096)
    print(status, report)
    t2 = datetime.now(UTC)
    dt = time_diff(t1, t2)
    for metric, value in params.items():
        full_frame_results[metric][i] = value
    print('Time taken = ' + repr(dt)+'s')
    total_time += dt

print('Total time taken = ' + repr(total_time) + 's; ' + str(total_time/len(results))+'s per frame')
print(full_frame_results)
```

Now we can compare the metrics from the frame center versus the full frame

```python
images = [path.basename(x) for x in image_list]
fig, axs = plt.subplots(4, 2)
plt.subplots_adjust(wspace=0.5)
xplt = 0
yplt = 0
for metric in results.keys():
    axs[yplt,xplt].plot(range(0,len(image_list),1), results[metric], 'g.', label='Centre')
    axs[yplt,xplt].plot(range(0,len(image_list),1), full_frame_results[metric], 'm*', label='Full frame')
    axs[yplt,xplt].set_ylabel(metric)
    if yplt == 3:
        axs[yplt,xplt].set_xlabel('Image')
        axs[yplt,xplt].set_xticks(range(0,len(image_list),1))
        axs[yplt,xplt].set_xticklabels(images, rotation=45)
    yplt += 1
    if yplt >= axs.shape[0]:
        xplt += 1
        yplt = 0
plt.show()
print('Green dots = frame center results; purple stars = full frame results')
```

The sky background and sky sigma metrics do not seem to give a reliable evaluation of the sky background.  

Looking at the output from the pipeline's starfind routine, DAOphot is used for the star detection, with a PSF fit performed from the pipeline's own code.  The values returned are in fact the background values for individual stars as determined from the PSF, with a fixed default PSF FWHM of 3pix. This was originally intended to measure the number of stars and the elongation of the PSF rather than give statistics on the image array itself. 

It looks as though the fixed-width PSF fit isn't producing reliable sky background estimates, so the next step is to explore alternatives.

```python
MAXLIN = 126000.0
sky = np.zeros(len(image_list))
image_data = {}
               
images = [path.basename(x) for x in image_list]
fig, axs = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.5, hspace=0.6)
xplt = 0
yplt = 0
for i,image_path in enumerate(image_list):
    imdata = fits.getdata(image_path)

    # Exclude the overscan regions
    ylim,xlim = imdata.shape
    imdata = imdata[100:ylim-100, 100:xlim-100]

    # Flattening the array makes the histogram considerably faster
    data = imdata.flatten()
    #percentile_lo = np.percentile(data, 5)
    #percentile_hi = np.percentile(data, 99)

    freq, bins, patches = axs[yplt,xplt].hist(data, bins=1000, log=True)
    [xmin,xmax,ymin,ymax] = axs[yplt,xplt].axis()
    axs[yplt,xplt].axis([0.0,5000.0,ymin,ymax])
    axs[yplt,xplt].set_xlabel('ADU')
    axs[yplt,xplt].set_ylabel('Frequency')
    axs[yplt,xplt].set_title(images[i].replace('-e91.fits',''))

    ibin = np.where(freq == freq.max())[0][0]
    print(images[i] + ' sky background = ' + str(bins[ibin]) + 'ADU')
    sky[i] = bins[ibin]
    image_data[images[i]] = imdata
    
    xplt += 1 
    if xplt >= 2:
        yplt += 1 
        xplt = 0
plt.show()
```

This metric is a more reliable measure of the sky background, and from the histograms, the presence of the structure in the background of coj1m003-fa19-20240722-0206 is clear from the 'rounding' of the peak in the histogram and the wide gap between the peak and the lowest values.  There is an unusually high proportion of pixels at intermediate-high sky values.  

The shorter peak at very low background valuyes is likely to be due to overscan/dead pixels.

In 'healthy' images, the historgram peak represents the median sky background with quite a narrow peak.  If the pixels covered by stars are excluded, then this peak will likely be clearer.  

```python
def gaussian(x, amp, mean, std):
    return amp * np.exp(-((x - mean) ** 2) / (2 * std ** 2))
```

```python
dx = 8.0 # pixels
dy = 8.0 # pixels
fig, axs = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.5, hspace=0.6)
xplt = 0
yplt = 0
for i,image_path in enumerate(image_list):
    imdata = image_data[path.basename(image_path)]
    daofind = DAOStarFinder(fwhm=3.0, threshold=5.0*np.sqrt(sky[i]))
    sources = daofind(imdata - sky[i])
    
    for j in range(0, len(sources), 1):
        x = int(sources['xcentroid'][j])
        y = int(sources['ycentroid'][j])
        xmin = int(max(x-dx, 1.0))
        xmax = int(min(x+dx, imdata.shape[1]))
        ymin = int(max(y-dy, 1.0))
        ymax = int(min(y+dy, imdata.shape[0]))
        #print(imdata[ymin:ymax, xmin:xmax], xmin, xmax, ymin, ymax)
        imdata[ymin:ymax, xmin:xmax] = sky[i]
        #print('AFTER: ',imdata[ymin:ymax, xmin:xmax])
        
    data = imdata.flatten()
    freq, bins, patches = axs[yplt,xplt].hist(data, bins=1000)

    #params, _ = curve_fit(gaussian, bins[0:1000], freq[0:1000])
    mean,std=norm.fit(data)
    
    # Extract the fit parameters
    #amp_fit, mean_fit, std_fit = params
    #print(amp_fit, mean_fit, std_fit)
    print(mean, std)
    
    [xmin,xmax,ymin,ymax] = axs[yplt,xplt].axis()
    axs[yplt,xplt].axis([0.0,5000.0,ymin,ymax])
    axs[yplt,xplt].set_xlabel('ADU')
    axs[yplt,xplt].set_ylabel('Frequency')
    axs[yplt,xplt].set_title(images[i].replace('-e91.fits',''))

    x1 = np.linspace(1.0, 5000.0, 100)
    y1 = norm.pdf(x1, mean, std)
    axs[yplt,xplt].plot(x1, y1, 'g-')
    
    xplt += 1 
    if xplt >= 2:
        yplt += 1 
        xplt = 0
plt.show()
```
